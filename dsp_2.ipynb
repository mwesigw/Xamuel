```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Fs = 5000  # Hz

# Define the time interval
Ts = 1 / Fs

# Define the range of n
n = np.arange(0, 100,Ts)

# Define the frequencies
Frequencies = [0,500, 2000, 3000, 4500]
```

```python
for Fo in Frequencies:
    # Generate the sampled signal
    x_n = np.sin(2 * np.pi * (Fo / Fs) * n)

    # Plot the signal
    plt.stem(n, x_n, label=f"Fo = {Fo} Hz")

# Add labels and legend
    plt.title("Sampled Signals for  Frequencies")
    plt.xlabel("n")
    plt.ylabel("x(n)")
    plt.legend()
    plt.grid(True)

# Show plot
    plt.show()
```

```python
Fs1= 50000  # in Hz

# Define the time period for sampling
Ts1 = 1 / Fs1

# Define the range of n
n_values = np.arange(0, 100,Ts1)

# Define the frequency Fo
Fo1 = 2000  # in Hz

# Calculate the sampled signal for the given Fo
x_n1 = np.sin(2 * np.pi * (Fo1 / Fs1) * n_values)
```

```python
 #Plot the sampled signal for x(n)
plt.figure(figsize=(8, 6))
plt.stem(n_values, x_n1, use_line_collection=True)
plt.title(f"Sampled Signal for Fs = {Fs1} Hz and Fo = {Fo1} Hz")
plt.xlabel('n')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
```
